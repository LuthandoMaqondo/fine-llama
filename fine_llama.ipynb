```python
# Run training
!autotrain llm \
    --train \
    --model "Llama-2-7b-hf/" \
    --project-name "fine-llama-2" \
    --log wandb \
    --data-path "data/" \
    --text-column text \
    --lr 2e-5 \
    --batch-size 4 \
    --epochs 3 \
    --block-size 1024 \
    --warmup-ratio 0.03 \
    --lora-r 16 \
    --lora-alpha 32 \
    --lora-dropout 0.05 \
    --weight-decay 0.0 \
    --gradient-accumulation 4 \
    --logging_steps 10 \
    # --fp16 \
    --use-peft \
    # --use-int4 \
    --merge-adapter \
    --push-to-hub \
    # --token <huggingface-token> \
    --repo-id "appimate/fine-llama-2"
```
